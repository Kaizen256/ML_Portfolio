```python
import json

with open("[2, 2, 6, 2].json", "r") as f:
    data = json.load(f)
```

```python
import plotly.graph_objects as go

epochs = list(range(1, int(len(data['train_loss']) + 1)))

fig = go.Figure()
fig.add_trace(go.Scatter(x=epochs, y=data['train_loss'], mode='lines', name='Train Loss'))
fig.add_trace(go.Scatter(x=epochs, y=data['val_loss'], mode='lines', name='Validation Loss'))

fig.update_layout(title='[2, 2, 6, 2] Training vs Validation Loss',
                  xaxis_title='Epoch',
                  yaxis_title='Loss',
                  template='plotly_white')
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=epochs, y=data['val_top1'], mode='lines', name='Top-1 Accuracy'))
fig.add_trace(go.Scatter(x=epochs, y=data['val_top5'], mode='lines', name='Top-5 Accuracy'))

fig.update_layout(title='[2, 2, 6, 2] Validation Top-1 and Top-5 Accuracy',
                  xaxis_title='Epoch',
                  yaxis_title='Accuracy',
                  template='plotly_white')
fig.show()
```
